```
from keras_segmentation.models.unet import resnet50_unet

model = resnet50_unet(n_classes=7 ,  input_height=1440 , input_width=1088  )
```

```
import os
os.environ["CUDA_VISIBLE_DEVICES"]="-1"    
import tensorflow as tf
```

```
model.train(
    verify_dataset=False,
    train_images =  "tmp/test/",
    train_annotations = "tmp/label/",
    checkpoints_path = "tmp/resnet50_unet_1" , epochs=4  
)
```

```
out = model.predict_segmentation(
    inp="tmp/test/image6.png",
    out_fname="/tmp/out.png"
)
```
